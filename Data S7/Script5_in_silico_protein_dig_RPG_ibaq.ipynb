## iBAQ

```python
def ibaqdict(name_df):
    name_dict = {}
    for row in range(name_df.shape[0]):
    #     count = 0
        key = name_df.loc[row,"Original_header"] #gene
#         print("key: ", key)
        if key not in name_dict.keys():
#             print("new gene")
            name_dict[key] = 0
        else: # new gene
#             print("not a new gene")
            count = name_dict[key]
            countnew = count + 1
            name_dict[key] = countnew
    # print(ibaq_dh)
    return name_dict
```

```python
ibaq_dh = ibaqdict(df_dh_350)
print("\n\n",ibaq_dh)
ibaq_zz = ibaqdict(df_zz_350)
print("\n\n",ibaq_zz)
ibaq_mm = ibaqdict(df_mm_350)
print("\n\n",ibaq_mm)
```

```python
# find unique peptides by gene

# df_mapping_dh = df_mapping.set_index('DH')
# df_mapping_dh['DH'] = df_mapping_dh.index
# print(df_mapping_dh.head())
```

```python
df_DH_peptides_gene = df_dh_350.set_index("Original_header")
df_MM_peptides_gene = df_mm_350.set_index("Original_header")
df_ZZ_peptides_gene = df_zz_350.set_index("Original_header")
```

```python
def df_or_ser(df_microbe):
    list_seq = []
    if isinstance(df_microbe, pd.DataFrame):
        list_seq = df_microbe['Sequence'].tolist()
    elif isinstance(df_microbe, pd.Series):
        list_seq = [df_microbe['Sequence']]
    return list_seq
```

```python
unique_pep_dict ={} # have the DH gene as the key and a list of unique peptides from Z-245 and Marburg

for g in range(df_mapping.shape[0]):
#     print(g)
    dh_gene = df_mapping.loc[g,"DH"]
    zz_gene = df_mapping.loc[g,"ZZ"]
    mm_gene = df_mapping.loc[g,"MM"]
    
    print(dh_gene)
    try:
        df_dh_gene = df_DH_peptides_gene.loc[dh_gene,:]
        list_seq_dh = df_or_ser(df_dh_gene)
#         print(list_seq_dh)
    except KeyError:
#         print("Not a real gene: ", dh_gene)
        list_seq_dh = []
#     print(zz_gene)
    try:
        df_zz_gene = df_ZZ_peptides_gene.loc[zz_gene,:]
        list_seq_zz = df_or_ser(df_zz_gene)
#         print(list_seq_zz)
    except KeyError:
#         print("Not a real gene: ", zz_gene)
        list_seq_zz = []

    
    print(mm_gene)
    try:
        df_mm_gene = df_MM_peptides_gene.loc[mm_gene,:]
        list_seq_mm = df_or_ser(df_mm_gene)
#         print(list_seq_mm)
    except KeyError:
#         print("Not a real gene: ", mm_gene)
        list_seq_mm = []

        
    list_seq_zz_mm = set(list_seq_mm + list_seq_zz )
#     print(list_seq_zz_mm)
    
    unique_pep_list = []
    
    for pep in list_seq_zz_mm:
#         print(pep)
        if pep in list_seq_dh:
            print("no need to add, already in DH: ", pep)
        else:
            print("add pep: ", pep)
            unique_pep_list.append(pep)
#     print("unique_pep_list: ", unique_pep_list)
    
    unique_pep_dict[dh_gene] = unique_pep_list
            
print(unique_pep_dict)
```

```python
protein_group_list_final= []
mm_gene_list_final = []
dh_gene_list_final = []
zz_gene_list_final = []
annotation_list_final = []
sequence_list_final = []


for g in proteingroup_dict:
    print(g)
    print(proteingroup_dict[g])
    protein_group_list_final.append(g)
    
    mm_gene = proteingroup_dict[g][0]
    dh_gene = proteingroup_dict[g][1]
    zz_gene = proteingroup_dict[g][2]
    print(mm_gene)
    print(dh_gene)
    print(zz_gene)
    mm_gene_list_final.append(mm_gene)
    dh_gene_list_final.append(dh_gene)
    zz_gene_list_final.append(zz_gene)

    try:
        # Add the DH protein without the C-terminal if the C-terminal is "Normal"
        dh_cterm = dh_cterm_dict[dh_gene]
        dh_cterm_len = len(dh_cterm)*-1
        dh_prot = dh_prot_dict[dh_gene]
        dh_prot_use = dh_prot[:dh_cterm_len] # removes the C-Terminal
        
        # for each protein go through the peptides
    except KeyError:
        print("DH C=terminal ends in K or R")
        try:
            dh_prot_use = dh_prot_dict[dh_gene] # Checks that this isn't a None gene
        except KeyError:
            print("Gene isn't in the genome, it is none: ", dh_gene)
            dh_prot_use = ""

    try:
        pep_to_add = unique_pep_dict[dh_gene]
        pep_str = " ".join(pep_to_add)
        merged_gene = dh_prot_use + pep_str
#     print(merged_gene)
    except KeyError:
        print("No list of unique peptides for this gene: ", dh_gene)
        merged_gene = dh_prot_use

    # Readd on a C-terminal
    try:
        cterm = df_cterm_dh.loc[dh_gene,"CTerm"]
        print("cterm in common, add", cterm)
        merged_gene_final = merged_gene + cterm
    except KeyError:
        print("Cterm NOT in common, try to add DHs")
        try:
            cterm = dh_cterm_dict[dh_gene]
            merged_gene_final = merged_gene + cterm
        except KeyError:
            print("DH Cterminal ends in K or R, so is already in the gene")
            merged_gene_final = merged_gene
    
    # removes spaces from final sequence
    merged_gene_final = merged_gene_final.replace(" ", "")
    print(merged_gene_final)
    sequence_list_final.append(merged_gene_final)
    
#     # adding the annotation
    try:
        dh_anno = dh_anno_dict[dh_gene]
        annotation_list_final.append(dh_anno)
        
    except KeyError:
        try:
            zz_anno = zz_anno_dict[zz_gene]
            annotation_list_final.append(zz_anno)
        except KeyError:
            mm_anno = mm_anno_dict[mm_gene]
            annotation_list_final.append(mm_anno)
```

```python
print(len(protein_group_list_final))
print(len(mm_gene_list_final))
print(len(dh_gene_list_final))
print(len(zz_gene_list_final))
print(len(annotation_list_final))
print(len(sequence_list_final))
```
